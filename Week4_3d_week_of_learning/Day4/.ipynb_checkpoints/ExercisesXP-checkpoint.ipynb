### Exercise 2: Creating a Line Plot for Temperature Variation

**Objective**: Create a simple line plot using Matplotlib that represents temperature variations over a week.

**Tasks**:
- Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]).
- Label the x-axis as “Day” and the y-axis as “Temperature (°F)”.
- Add a title to the plot.
- Display the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (3,3))

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
tempreture = [72, 74, 76, 80, 82, 78, 75]

plt.plot (days, tempreture, color='lightcoral', linestyle='dotted', linewidth=2.5, marker='o', markerfacecolor='darkturquoise', markersize=6)
plt.title('Tempreture by Days of the Week')
plt.xlabel('Tempreture(F)')
plt.ylabel('Day')
plt.show()
```

### Exercise 3: Visualizing Monthly Sales with a Bar Chart

**Objective**: Generate a bar chart using Matplotlib to visualize monthly sales data for a retail store.

**Tasks**:
- Create a list of sales values for each month (e.g., [5000, 5500, 6200, 7000, 7500]).
- Label the x-axis as “Month” and the y-axis as “Sales Amount ($)”.
- Add a title to the bar chart.
- Display the plot.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
amount = [3000, 3600, 4100, 4700, 5000, 5500, 6200, 7000, 7500, 7600, 7600, 8000]

figure, axes = plt.subplots(figsize = (10,3))
axes.bar(months, amount, color = ['royalblue', 'lightseagreen', 'pink', 'darkturquoise', 'moccasin', 'lightcoral'], edgecolor='grey', hatch='')
axes.set_title('Average amout for months')
axes.set_xlabel('Month')
axes.set_ylabel('Sales Amount ($)')
```

### Exercise 4: Visualizing the Distribution of CGPA

**Objective**: Create a histogram to visualize the distribution of students’ CGPA.

**Dataset Overview**: Assume the CGPA data is categorized into ranges and loaded in a DataFrame named df.

**Tasks**:
- Import necessary libraries.
- Use Seaborn’s histplot to create a histogram of the CGPA categories.
- Customize the histogram with a specific color and add a title.
- Display the plot.

```python
import seaborn as sns
```

```python
students_d = pd.read_csv('Student Mental health.csv')
```

```python
students_d.info()
```

```python
students_d['What is your CGPA?'] = students_d['What is your CGPA?'].str.strip() # removing spaces
```

```python
students_d['What is your CGPA?'].value_counts()
```

```python
students_CGPA = students_d.set_index('What is your CGPA?') # checking what is wrong with column
students_CGPA.loc['2.00 - 2.49']
```

```python
students_d['CGPA_num'] = students_d['What is your CGPA?'].str.split('-').str[0].astype(float) # take only upper line
```

```python
order = [
    "0 - 1.99",
    "2.00 - 2.49",
    "2.50 - 2.99",
    "3.00 - 3.49",
    "3.50 - 4.00"
] # new categories

students_d["CGPA_cat"] = pd.Categorical(
    students_d["What is your CGPA?"].str.strip(), # cleaning spaces
    categories=order, # defining categories
    ordered=True # telling that categories have order
)
```

```python
len(students_d) # checking that number of rows are similar to students_d["CGPA_cat"]
```

```python
len(students_d["CGPA_cat"])
```

```python
plt.figure(figsize=(8, 3))
sns.histplot(students_d, x = 'CGPA_cat', color=sns.color_palette("flare")[4]) 
plt.title('The CGPA level among students')
plt.xlabel('CGPA level')
plt.ylabel('Quantity')
plt.show()
```

### Exercise 5: Comparing Anxiety Levels Across Different Genders

**Objective**: Use a bar plot to compare the proportion of students experiencing anxiety across different genders.

**Dataset Overview**: The dataset includes columns: ‘Do you have Anxiety?’ and ‘Choose your gender’.

**Tasks**:
- Import necessary libraries.
- Use Seaborn to create a bar plot comparing anxiety levels across genders from the dataset df.
- Customize the plot with an appropriate color palette and add a title.
- Display the plot.

```python
students_d['Do you have Anxiety?'].value_counts() # checking the type of answers
```

```python
students_d['Anxiety'] = students_d['Do you have Anxiety?'].map({
    'No': 0,
    'Yes': 1
}) #creating new column only with anxiety
```

```python
sns.barplot(x='Choose your gender', y='Anxiety', data=students_d, palette='Purples', hue='Choose your gender')
plt.title('The proportion of students experiencing anxiety across different genders')
plt.xlabel('Gender')
plt.ylabel('Proportion with Anxiety')
plt.ylim(0, 1)
plt.show()
```

### Exercise 6: Exploring the Relationship Between Age and Panic Attacks

**Objective**: Create a scatter plot to explore the relationship between students’ age and the occurrence of panic attacks.

**Dataset Overview**: The dataset includes columns: ‘Age’ and ‘Do you have Panic Attacks?’.

**Tasks**:
- Import necessary libraries.
- Convert panic attack responses to numeric values (e.g., Yes=1, No=0).
- Use Seaborn’s scatterplot to create a scatter plot with ‘Age’ on the x-axis and numeric panic attack responses on the y-axis.
- Customize the plot to improve readability by adding labels, a title, and adjusting point styles.
- Display the plot.

```python
students_d.head()
```

```python
students_d['Panic_attack'] = students_d['Do you have Panic attack?'].map({
    'No': 0,
    'Yes': 1
}) #creating new column only with panic attack
```

```python
sns.scatterplot(x='Age', y='Panic_attack', data=students_d, alpha=0.7, s=100)
plt.title('The Relationship Between Students’ Age and the Occurrence of Panic Attacks')
plt.xlabel('Age')
plt.ylabel('Panic Attack')
plt.show()
```

```python
sns.barplot(data = students_d, x= 'Age', y ='Panic_attack', palette = ['royalblue', 'lightseagreen', 'pink', 'darkturquoise', 'moccasin'])
plt.title('The Relationship Between Students’ Age and the Occurrence of Panic Attacks')
plt.xlabel('Age')
plt.ylabel('Panic Attack') # it looks a little bit better than scatter
```
