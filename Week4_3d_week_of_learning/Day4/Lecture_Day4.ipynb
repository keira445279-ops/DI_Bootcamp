## Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Anatomy of a Matplotlib

Figure (the whole canvas)

    | - Axes(the plot area with data)
       | - X - axis
       | - Y - axis
       | - Title
       | - Legend
       | - Data (lines, points, bars and etc.)

#### Ways to create plots:
**1. simple way: plt.plot()**<br>
**2. more complex: figure, axes = plt.subplots()**

Pattern:

- plt.plot(x, y) # create plot
- plt.title()
- plt.xlabel()
- plt.ylabel()
- plt.show() # Display the plot

```python
# See thr trend of the mpg change across time

mpg.groupby('model_year')['mpg'].mean()
```

```python
mpg_by_year = mpg.groupby('model_year')['mpg'].mean()

years = mpg_by_year.index # X-axis
average_mpg = mpg_by_year.values # Y-axis
```

```python
# LINE PLOT

plt.figure(figsize = (3,3))
plt.plot(years, average_mpg) # create plot
plt.title('Average MPG Over Time')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.show() # Display the plot
```

```python
# SCATTER PLOT: IS GOOD GOR CORRELATIONS/RELATIONSHIP BETWEEN COLUMNS

horsepower = mpg['horsepower']
mpg_data = mpg['mpg']

plt.figure(figsize = (10,3))
plt.scatter(horsepower, mpg_data)
plt.xlabel('Horsepower')
plt.ylabel('MPG')
```

```python
# BAR PLOT
# See the average MPG per Origin

mpg_by_origin = mpg.groupby('origin')['mpg'].mean()
mpg_by_origin
```

```python
origins = mpg_by_origin.index
average_mpg = mpg_by_origin.values

plt.figure(figsize = (5,3))
plt.bar(origins, average_mpg)
plt.xlabel('Origin')
plt.ylabel('Average MPG')
```

### More Complex Way: figure, axes = plt.subplots()

Pattern for single plot:

- figure, axes = plt.subplots()
- axes.set_title('....')
- axes.set_xlabel()
- axes.set_ylabel()

plt.show()

```python
figure, axes = plt.subplots(figsize = (8,3)) # figsize = (x, y)
# a single row and a single column axes, and it can be accessed directly (without indexing)
all_mpgs = mpg['mpg']
axes.hist(all_mpgs)
axes.set_title('Distribution of Cars MPGs')
axes.set_xlabel('MPG')
axes.set_ylabel('Quantity')

plt.show()
```

```python
# how to convert to numeric

origin_map = {'usa': 0, 'japan': 1, 'europe': 2}

def convert_origin_to_number (origin: str) -> int:
    return origin_map['origin']
    
mpg['origin_num'] = origin.apply(convert_origin_to_number)
```

```python
origins = mpg['origin'].unique()
# take all european cars then groupby model year and calculate the average mpg

european_cars_map = mpg['origin'] == 'europe'
usa_cars_map = mpg['origin'] == 'usa'
japanese_cars_map = mpg['origin'] == 'japan'

european_cars = mpg[european_cars_map] 
usa_cars = mpg[usa_cars_map] 
japanese_cars = mpg[japanese_cars_map] 

mpg_change_europe = european_cars.groupby('model_year')['mpg'].mean()
mpg_change_usa = usa_cars.groupby('model_year')['mpg'].mean()
mpg_change_japan = japanese_cars.groupby('model_year')['mpg'].mean()
```

```python
figure, axes = plt.subplots(nrows = 3, ncols = 1, figsize = (10, 6))

# plt.tight_layout()

x_europe = mpg_change_europe.index
y_europe = mpg_change_europe.values

axes[0].plot(x_europe,y_europe) # 0 because we have one column
axes[0].set_title('MPG Trend for Europe')
axes[0].set_xlabel('Model Year')
axes[0].set_ylabel('Average MPG')

x_usa = mpg_change_usa.index
y_usa = mpg_change_usa.values

axes[1].plot(x_usa,y_usa) # 0 because we have one column
axes[1].set_title('MPG Trend for USA')
axes[1].set_xlabel('Model Year')
axes[1].set_ylabel('Average MPG')

x_japan = mpg_change_japan.index
y_japan = mpg_change_japan.values

axes[2].plot(x_japan,y_japan) # 0 because we have one column
axes[2].set_title('MPG Trend for Japan')
axes[2].set_xlabel('Model Year')
axes[2].set_ylabel('Average MPG')


# figure.supxlabel("Model Year")   #     x-
# plt.subplots_adjust(hspace=0.4)
lt.tight_layout(rect=[0, 0.03, 1, 1]) #  

plt.show()
```
