### Exercise 5: Creating a Synthetic Product Catalog

- Ensure that the Faker library is installed and imported into your Python environment.

- Use Pandas to create a DataFrame from the generated data.

```python
from faker import Faker
fake = Faker()
```

- Create a dataset of 500 products. Each product should have a unique ID, name, description, and price.
- Use Pandas to create a DataFrame from the generated data.

```python
import random
```

```python
num_records = 500

products = []
for _ in range(num_records):
    product_id = fake.unique.random_int(min = 100000, max = 999999)  
    product_name = fake.word().capitalize() + " " + random.choice ([
        "Lamp", "Chair", "Table", "Notebook", "Bottle", "Backpack", "T-shirt",
        "Shoes", "Watch", "Phone Case", "Headphones", "Mug"
    ])
    description = fake.sentence(nb_words = 12)
    price = round(random.uniform(5, 500), 2)

    products.append({'ProductID': product_id,
                'Name': product_name,
                'Description': description,
                'Price': price
                })

products_d = pd.DataFrame(products)
products_d.head(10)
```
