```python
from scipy.ndimage import rotate
def rotate_image_30_degrees(original_image):
    return rotate(original_image, 30, reshape=False) #complete it
# original_image  NumPy array
# 30  rotation angle
# reshape=False  size of the picture will not change
```

```python
from PIL import Image, ImageOps
```

```python
# from left to right
flipped_img = original_image.transpose(Image.FLIP_LEFT_RIGHT)

# save the result
flipped_img.save("flipped_horizontal.png")

# from top to bottom
flipped_img = original_image.transpose(Image.FLIP_TOP_BOTTOM)

# save the result
flipped_img.save("flipped_vertical.png")
```
