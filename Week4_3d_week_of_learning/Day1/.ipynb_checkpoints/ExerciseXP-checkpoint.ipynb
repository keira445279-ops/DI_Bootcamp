### Exercise 1: Introduction to Data Analysis

- **What is data analysis?**
Data analysis is the process of collecting, cleaning, transforming, and interpreting data to discover meaningful patterns, answer questions, and support decision-making. It involves using statistical techniques, analytical methods, and visualization tools to extract insights from raw information.
- **Why is data analysis important in modern contexts?**
Data analysis is important because modern organizations generate vast amounts of data from digital interactions, sensors, business operations, and online platforms. Analyzing this data helps companies and institutions:
    - make evidence-based decisions
    - improve efficiency and reduce costs
    - identify trends and predict future outcomes
    - personalize services and products
    - detect risks, errors, or anomalies early<br>
In a data-driven world, organizations that use data effectively gain a competitive advantage.
- **List and describe three areas where data analysis is applied today.**
There are three areas where data analysis is applied today:
    - *HealthCare* - Data analysis is used to track patient health, predict disease risks, improve treatment effectiveness, and optimize hospital operations. For example, analyzing medical records and sensor data helps detect patterns that may indicate early stages of illness.
    - *Finance* - Banks and financial institutions use data analysis for fraud detection, credit scoring, investment prediction, and risk assessment. Algorithms can monitor transactions in real time and identify unusual activity.
    - *Buisness and Marketing* - Companies analyze customer behavior, sales data, and market trends to improve product recommendations, optimize pricing, target advertisements, and forecast demand. This helps increase revenue and enhance customer experience.

### Exercise 5: Basic Data Analysis with Google Colab

```python
mean_sepa_l = iris_data['SepalLengthCm'].mean()
median_sepa_l = iris_data['SepalLengthCm'].median()
mode_sepa_l = iris_data['SepalLengthCm'].mode()[0]

print('SepalLengthCm - Mean: ', mean_sepa_l)
print('SepalLengthCm - Median: ', median_sepa_l)
print('SepalLengthCm - Mode: ', mode_sepa_l)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(iris_data['SepalLengthCm'], kde=False, bins=15, color='cornsilk')

plt.axvline(mean_sepa_l, color='darksalmon', linestyle='--', label=f'Mean: {mean_sepa_l:.2f}')
plt.axvline(median_sepa_l, color='darkviolet', linestyle='--', label=f'Mean: {median_sepa_l:.2f}')
plt.axvline(mode_sepa_l, color='teal', linestyle='--', label=f'Mean: {mode_sepa_l:.2f}')

plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')
plt.show()
```

### Exercise 6: Basic Observation Skills in Data Analysis

```python
sleep_data = pd.read_csv('Time Americans Spend Sleeping.csv')
```

```python
sleep_data.info()
```

For this dataset we can use **three specific types** of analysis:
- trend analys
- group comparison
- correlation analys

##### Trend analyse

The dataset supports multiple types of trend analysis:
- Sleep duration over time (Year) â€” to see how sleep habits have changed.
- Sleep duration over time by Type of Days: how weekend sleep increase over the tear, how holiday sleep change

##### Group comparison

This dataset is very well suited for group comparison:
- Age Group - we can compare average sleep duration across different age groups to see whether younger or older people tend to sleep more.
- Sex - allows comparing sleep patterns between males and females.
- Type of Days - enables analysis of whether people sleep more on weekends and holidays than on weekdays.
- Activity - if activity type affects sleep, we can compare the groups.

##### Correlation analyse

Although most categorical variables in the dataset cannot be used directly for correlation analysis, it is still possible to perform a meaningful correlation-based exploration.
- Age Group vs Sleep Duration<br>
Even though Age Group is categorical, we can calculate the average sleep duration for each age category and examine whether sleep changes systematically with age.
This does not produce a numeric correlation coefficient but still shows a trend-like relationship.

- Type of Days vs Sleep Duration<br>
By comparing average sleep time for weekdays, weekends, and holidays, we can identify whether the type of day is associated with differences in sleep duration.
