As an example, I will refer to a real project I worked on at my previous job.<br> 
A client wanted to launch a new cold medicine and asked us to test several packaging options to understand how well each design would stand out among competitors on the pharmacy shelf, as well as to evaluate the key KPIs for each option: how attractive the packaging is to consumers and how much it motivates them to purchase the product.

**Step 2: Analyze the role of data analysis in the chosen story. Pay attention to what data was analyzed, the methods used, and the outcomes or decisions driven by the analysis.**

This study was primarily based on group comparison.<br>
The entire sample was divided into three main groups because there were three packaging concepts.<br> 
Each group evaluated one new packaging design along with competitors’ packaging.<br> 
Then, the new designs were compared with each other, and each new design was compared with competitors and the overall competitor benchmark.<br> 
This allowed us to identify which new packaging design stood out both against competitors and among the new options themselves.

**Step 3: Ask yourself how the data analysis impacted the situation. Consider questions like: What would have been different without data analysis? How did data analysis contribute to solving a problem or making a decision?**

Why was data analysis so critical in this case?<br> 
The main goal was to avoid choosing the wrong marketing strategy.<br> 
Without conducting research and simply selecting one of the three packaging options at random, the manufacturer would not know the product’s strengths and weaknesses compared with competitors. As a result, the chosen packaging might fail to catch consumers’ attention, get lost among other products on the shelf, and ultimately lead to low sales. Data analysis made it possible to avoid this outcome and choose the packaging option most relevant and appealing to consumers.

This is a simple example from the marketing field that clearly shows why a product should be tested before entering the real market.<br> 
It is impossible to predict with 100% certainty that one packaging design will be equally attractive to all consumers.<br> 
Data analysis helps determine which packaging has stronger KPIs and what improvements need to be made to ensure that the product sells well and the company achieves profit.
