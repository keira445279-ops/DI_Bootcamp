# Selecting, Filtering, Masking

```python
type(titanic['age'])
```

```python
age = titanic['age']
```

```python
sum(age > 30)
```

```python
sum (age < 30)
```

```python
titanic[['sex', 'age']]
```

```python
sex_and_age_titanic = titanic[['sex', 'age']]
```

```python
sex_and_age_titanic.isna().sum()
```

```python
sex_and_age_titanic['sex'].value_counts()
```

```python
sex_and_age_titanic['sex'].value_counts(normalize=True)
```

```python
# groupby - groups the data by a specific value (Female/Male)
titanic.groupby ('sex')['age'].mean()
```

```python
sex_and_age_titanic[sex_and_age_titanic['age'] < 14]
```

```python
children_sex_age_titanic = sex_and_age_titanic[sex_and_age_titanic['age'] < 14]
```

```python
children_sex_age_titanic['sex'].value_counts()
```

```python
young_sex_age_titanic = sex_and_age_titanic[(sex_and_age_titanic['age'] > 14) & (sex_and_age_titanic['age'] < 30) ]
```

```python
young_sex_age_titanic['sex'].value_counts()
```

```python
bins = [0, 14, 20, 30, 60, 120] # age categories
labels = ['child', 'teen', 'young adult', 'adult', 'senior']
```

```python
sex_and_age_titanic['age_group'] = pd.cut(sex_and_age_titanic['age'], bins = bins, labels = labels)
```

```python
sex_and_age_titanic['age_group'].value_counts()
```
