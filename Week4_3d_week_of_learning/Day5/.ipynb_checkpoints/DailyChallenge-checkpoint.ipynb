## 2. Data Visualization with Matplotlib:
- Create an interactive line chart to show sales trends over the years.
- Build an interactive map to visualize sales distribution by country.

```python
store_us['Year'] = store_us['Order Date'].dt.year # converting data only for year
```

```python
sales_by_year = store_us.groupby('Year')['Sales'].sum().reset_index() # grouping sales sum by year and converting it to df

sales_by_year = sales_by_year.sort_values(by='Year') # sorting by year
```

```python
# Creating lineplot Matplotlib, using mpld3
fig, ax = plt.subplots(figsize=(10, 6)) 

ax.plot(sales_by_year['Year'], sales_by_year['Sales'], 
        marker='o', linestyle='-', color='skyblue', label='Sales value') #building line chart with matplotlib, customizing markers on the line, color of the line

# Title and axis
ax.set_title('Sales Trends Over The Years', fontsize=25, pad = 20) # making title with suituble font size and indent
ax.set_xlabel('Year', fontsize=17, labelpad = 15) # making title of x-axis with suituble font size and indent
ax.set_ylabel('Sales (USD)', fontsize=17, labelpad = 15) # making title of y-axis with suituble font size and indent
ax.grid(True, linestyle='--', alpha=0.7) # making vertical lines on chart to make it more readable
ax.legend() # showing the legend

mpld3.display(fig) # using mpld3, because only this library allows to lauch chart with interactive tools
```

```python
sales_by_state = store_us.groupby('State')['Sales'].sum().reset_index() # grouping sales sum by states and converting this data to df
```

```python
# making a new table with short names of states names to build an interractive map

state_to_code = {
    'Alabama':'AL', 'Alaska':'AK', 'Arizona':'AZ', 'Arkansas':'AR',
    'California':'CA', 'Colorado':'CO', 'Connecticut':'CT', 'Delaware':'DE',
    'Florida':'FL', 'Georgia':'GA', 'Hawaii':'HI', 'Idaho':'ID',
    'Illinois':'IL', 'Indiana':'IN', 'Iowa':'IA', 'Kansas':'KS',
    'Kentucky':'KY', 'Louisiana':'LA', 'Maine':'ME', 'Maryland':'MD',
    'Massachusetts':'MA', 'Michigan':'MI', 'Minnesota':'MN', 'Mississippi':'MS',
    'Missouri':'MO', 'Montana':'MT', 'Nebraska':'NE', 'Nevada':'NV',
    'New Hampshire':'NH', 'New Jersey':'NJ', 'New Mexico':'NM', 'New York':'NY',
    'North Carolina':'NC', 'North Dakota':'ND', 'Ohio':'OH', 'Oklahoma':'OK',
    'Oregon':'OR', 'Pennsylvania':'PA', 'Rhode Island':'RI', 'South Carolina':'SC',
    'South Dakota':'SD', 'Tennessee':'TN', 'Texas':'TX', 'Utah':'UT',
    'Vermont':'VT', 'Virginia':'VA', 'Washington':'WA', 'West Virginia':'WV',
    'Wisconsin':'WI', 'Wyoming':'WY'
}

sales_by_state['Code'] = sales_by_state['State'].map(state_to_code) # creating a new column in sales_by_state map
```

```python
import plotly.express as px
```

```python
# Interactive map
fig = px.choropleth(
    sales_by_state,
    locations='Code',          
    locationmode='USA-states',  
    color='Sales',              
    scope='usa',                
    color_continuous_scale="Plasma",  
    labels={'Sales':'Total Sales'}
)

# Title
fig.update_layout(title_text='Interactive Sales Distribution by State')

fig.show()
```

## 3. Data Visualization with Seaborn:

- Use Seaborn to generate a bar chart showing top 10 products by sales.
- Create a scatter plot to analyze the relationship between profit and discount.<br>
 *Products with higher discounts tend to generate lower profit. This is mainly due to technology and office supplies.*

```python
top10_products = store_us.groupby('Product Name')['Sales'].sum().sort_values(ascending = False) # grouping sum sales by products names and sorting this data from max to min
top10_products = top10_products.reset_index().head(10) # showing first 10 rows
top10_products
```

```python
# making a barplot for sum sales for each product from top-10

plt.figure(figsize = (15,6)) # changing the size of the chart
sns.barplot(data = top10_products, x = 'Product Name', y = 'Sales', hue = 'Product Name', palette = 'Set2') # building a bar chart with seaborn and customize the color
plt.title('TOP-10 Products by Sales', fontsize=20, pad = 20) # making the title of chart with suitable font size and indent
plt.xlabel('Products', fontsize=17) # making the title of chart with suitable font size
plt.ylabel('Sales', fontsize=17, labelpad=20) # making the title of chart with suitable font size and indent
plt.xticks(rotation = 90) # making rotation fo x_axis lables to make them more readeble

plt.show() # lauching the chart
```

```python
# making the scatterplot to find out about elationship between profit and discount

plt.figure(figsize = (15,5)) # changing the size of the chart

sns.scatterplot(data = store_us, x = 'Profit', y = 'Discount', hue = 'Category') # building scatter chart with seaborn
plt.title('Relationship between Profit and Discount', fontsize = 25, pad = 20) # making the title of chart with suitable font size and indent
plt.xlabel('Profit level', fontsize = 17, labelpad = 20) # making x-axis with suitable font size and indent
plt.ylabel('Discount level', fontsize = 17, labelpad = 20) # making y-axis with suitable font size and indent

plt.show() # lauching the chart
```

## 4. Comparative Analysis:
- Compare the insights gained from Matplotlib and Seaborn visualizations.
- Document your observations about the ease of use and effectiveness of both tools.

**Comparative Analysis: Matplotlib vs Seaborn**

**Matplotlib**

**Advantages**:

- A very flexible, low-level visualization library that allows full control over every element of a plot.

- Supports the creation of almost any type of chart.

- Suitable for detailed customization, including colors, labels, fonts, grids, and more.

**Disadvantages**:

- Requires more code to produce visually appealing plots.

- Default plot styles look less modern.

- Many aesthetic settings (e.g., style, font size, legends) must be configured manually, which takes additional time.

**Seaborn**

**Advantages**:

- Built on top of Matplotlib but provides a higher-level, more convenient interface.

- Produces attractive, polished visualizations by default.

- Easily creates complex plots such as boxplots, violin plots, heatmaps, and pairplots.

- Integrates seamlessly with pandas DataFrames.

**Disadvantages**:

- Offers less fine-grained control compared to Matplotlib.

- Customization options are limited in certain advanced or non-standard use cases.

- Some styles and parameters may conflict with underlying Matplotlib settings.


**Conclusion**

- Matplotlib is ideal for detailed customization and highly specialized or complex plots.

- Seaborn is excellent for fast, convenient, and aesthetically pleasing visualizations, particularly during exploratory data analysis (EDA).

- The best approach is often to combine both libraries: use Seaborn for high-level visualizations and Matplotlib for fine-tuning and customization.
