## 2. What is the difference between New York and California in terms of sales and profit?<br> (Compare the total sales and profit between New York and California.)

In California, total sales are higher than in New York. This could be related to the higher number of orders in California. Despite this, profit in both states is quite high, but California shows a slightly higher total profit.

```python
store_ca_ny = superstore[superstore['State'].isin(['California', 'New York'])] # filtering by NY and CA
```

```python
store_ca_ny.head()
```

```python
ca_kpi_sum = store_ca_ny.groupby('State')[['Sales', 'Profit']].sum() # grouping sales and profit sum by states
ca_kpi_sum
```

```python
# creating barplots

figure, axes = plt.subplots(ncols = 2, nrows = 1, figsize = (15,3)) 

palette = sns.color_palette('Set2')

sns.barplot(data = ca_kpi_sum, x = 'State', y = 'Sales', ax = axes[0], hue = 'State', palette = palette[:2])
axes[0].set_title('Difference Between New York and California In Terms Of Sales')
axes[0].set_ylabel('Total Sales')

sns.barplot(data = ca_kpi_sum, x = 'State', y = 'Profit', ax = axes[1], hue = 'State', palette = palette[3:5])
axes[1].set_title('Difference Between New York and California In Terms Of Profit')
axes[1].set_ylabel('Total Profit')

plt.show()
```

## 4. Are there any differences among states in profitability?

- California, West Virginia, and Wyoming have total profit above the average, but it is not negative.
- Negative profitability is observed in Arizona, Colorado, Florida, Illinois, North Carolina, Ohio, Oregon, Pennsylvania, Tennessee, and Texas.
- Other states have profitability at or above the average.

```python
profit_all_states = superstore.groupby('State')[['Sales', 'Profit']].sum() # grouping sales, profit sum by states
```

```python
profit_all_states['Profitability'] = (profit_all_states['Profit']/profit_all_states['Sales']) 
```

```python
mean_value = profit_all_states['Profitability'].mean()

plt.figure(figsize = (15,5))
sns.barplot(data = profit_all_states, x = 'State', y = 'Profitability', hue = 'State', legend = False)
plt.axhline(mean_value, color='red', linestyle='--', linewidth=1) # Profit mean
plt.xticks(rotation=90)
plt.title('Differences Among States In Profitability')
plt.ylabel('Total Profit')
```

## 5. The Pareto Principle, also known as the 80/20 rule, is a concept derived from the work of Italian economist Vilfredo Pareto. It states that roughly 80% of the effects come from 20% of the causes. For instance, identifying the top 20% of products that generate 80% of sales or the top 20% of customers that contribute to 80% of profit can help in prioritizing efforts and resources. This focus can lead to improved efficiency and effectiveness in business strategies. Can we apply Pareto principle to customers and Profit ? (Determine if 20% of the customers contribute to 80% of the profit.)

We can apply Pareto Principle to customers and Profit, because 20% of Customers contribute 81% of the profit.

```python
superstore['Customer ID'].shape
```

```python
superstore['Customer ID'].unique().shape # checking unique customers
```

```python
cust_profit = superstore.groupby('Customer ID')['Profit'].sum().sort_values(ascending=False) # grouping profit sum by customers
```

```python
cust_profit_total = cust_profit.sum() # total profit
```

```python
customer_profit_cum = cust_profit.cumsum() / cust_profit_total # cumulative profit
```

```python
n_customer = len(cust_profit) # customers number
top_20_perc = int(n_customer * 0.2) # 20% from customers number

top_customers = cust_profit.iloc[:top_20_perc] # choosing 20% of customers

top_customers_profit_share = top_customers.sum()/cust_profit_total
top_customers_profit_share
```

## 8. Plot the Cumulative curve in Sales by Customers.<br> Can we apply Pareto principle to customers and Sales?

We couldn`t apply Pareto principle to customers and sales

```python
cust_sales_sum = superstore.groupby('Customer ID')['Sales'].sum().sort_values(ascending = False) # grouping sales sum by customers
cust_sales_total = cust_sales_sum.sum() # counting total sum for all customers
```

```python
cust_sales_cum = cust_sales_sum.cumsum()/cust_sales_total # cumulative sales
```

```python
# n_customers = len(cust_sales_sum) # number of customers
# top_20_perc_cust = int(n_customers * 0.2) # 20% from customers

# top_20_perc_cust_sales = cust_sales_sum.iloc[:top_20_perc_cust] # choosing 20% from customers 

# cust_sales_par = top_20_perc_cust_sales.sum()/cust_sales_total
# cust_sales_par
```

```python
plt.figure(figsize=(10,6))

plt.plot(cust_sales_cum.index, cust_sales_cum.values, color = 'salmon')
plt.title('Pareto Curve for Customers and Sales')
plt.xlabel('Customer ID')
plt.ylabel('Cumulative Share of Sales')

plt.axhline(0.8, color='red', linestyle='--', label='80% of Sales')
plt.axvline(int(len(cust_sales_cum)*0.2), color='green', linestyle='--', label='20% of Customers')


plt.xticks(cust_sales_cum.index[::25], rotation=90)

plt.show()
```
