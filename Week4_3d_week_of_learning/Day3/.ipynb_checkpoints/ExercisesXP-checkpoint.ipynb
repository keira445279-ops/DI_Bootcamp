### Exercise 3: Feature Engineering

**Instructions**
- Create new features, such as Family Size from SibSp and Parch, and Title extracted from the Name column.
- Convert categorical variables into numerical form using techniques like one-hot encoding or label encoding.
- You will encode new categorical features (like Title) here, but do not scale numerical features yet â€” that will come after outlier handling.
- **Hint**: Utilize Pandas for data manipulation and scikit-learnâ€™s preprocessing module for encoding.

```python
train_d.head()
```

```python
train_d['FamilySize'] = train_d['SibSp'] + train_d['Parch'] + 1 # creating new feature 
```

```python
train_d['Title'] = train_d['Name'].str.extract(r",\s*([^\.]+)\.") # creating new feature with method str.extract
```

```python
train_d.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# one hot encoding for FamilySize
ohe = OneHotEncoder(sparse_output=False, drop='first') 
family_encoded = ohe.fit_transform(train_d[['FamilySize']])
family_encoded_train_d = pd.DataFrame(family_encoded, columns=ohe.get_feature_names_out(['FamilySize']))
train_d = pd.concat([train_d, family_encoded_train_d], axis=1)

# encoding by lables for Title
le = LabelEncoder()
train_d['Title'] = le.fit_transform(train_d['Title'])
```

```python
train_d[['Title'] + list(family_encoded_train_d.columns)].head() # checking new columns
```

### Exercise 4: Outlier Detection and Handling

**Goal**: Detect and cap or transform outliers in columns like Fare and Age.

1. Visualize distributions using boxplots or histograms to identify potential outliers.
2. Use IQR or Z-score methods to detect them.
3. Handle outliers with:
    - Quantile capping (e.g. 0.98)
    - Log transformation
    - Row removal
      
4. Compare the dataset before and after treatment.

ðŸ“Œ **Note**: Small differences between 0.98 and 0.99 quantiles are normal when extreme values are rare or far apart. Use df.quantile() to explore and choose thresholds empirically, backed by visualization.

```python
import matplotlib.pyplot as plt
```

```python
# Boxplot for Age
plt.figure(figsize=(3,2))
plt.boxplot(train_d['Age'].dropna())
plt.title('Boxplot of Age')
plt.ylabel('Age')
plt.show()

# Boxplot for Fare
plt.figure(figsize=(3,2))
plt.boxplot(train_d['Fare'].dropna())
plt.title('Boxplot of Fare')
plt.ylabel('Fare')
plt.show()
```

```python
# IQR for Age
Q1 = train_d['Age'].quantile(0.25)
Q3 = train_d['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_age = Q1 - 1.5 * IQR
upper_age = Q3 + 1.5 * IQR
```

```python
# IQR for Fare

Q1 = train_d['Fare'].quantile(0.25)
Q3 = train_d['Fare'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
train_data = train_d[((train_d['Age'] >= lower_age) & (train_d['Age'] <= upper_age)) & ((train_d['Fare'] >= lower_bound) & (train_d['Fare'] <= upper_bound))]
```

```python
len(train_d) # cheking the length before IQR
```

```python
len(train_data) # cheking the length after IQR
```

```python
# Quantile caping
# Cap Fare at 98th percentile
upper_fare_cap = train_d['Fare'].quantile(0.98)
train_d['Fare_capped'] = train_d['Fare'].apply(lambda x: min(x, upper_fare_cap))

# Cap Age at 98th percentile
upper_age_cap = train_d['Age'].quantile(0.98)
train_d['Age_capped'] = train_d['Age'].apply(lambda x: min(x, upper_age_cap))
```

```python
# Log transformation

import numpy as np
train_d['Fare_log'] = np.log1p(train_d['Fare'])  
```

```python
# Compare the dataset before and after treatment

plt.figure(figsize=(4,3))
plt.boxplot(train_d['Fare_log'].dropna())
plt.title('Log-transformed Fare')
plt.show()
```

### Exercise 7: Data Transformation for Age Feature
