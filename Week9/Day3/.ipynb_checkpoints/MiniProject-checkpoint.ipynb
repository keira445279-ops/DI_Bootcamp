```python
import pprint
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import chromedriver_autoinstaller

# STEP 1: Setup and open the browser
chromedriver_autoinstaller.install()
driver = webdriver.Chrome()

# Target URL for Shared Hosting plans
url = "https://www.inmotionhosting.com/shared-hosting"
driver.get(url)

# STEP 2: Wait for the page to load completely
print("Loading page... please wait.")
time.sleep(10) 

# STEP 3: Pass the page source to BeautifulSoup for parsing
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all product cards using the common class name
cards = soup.find_all('div', class_=lambda x: x and 'imh-rostrum-card' in x)

plans_list = []

for card in cards:
    # Extract the Plan Name (h3 tag)
    name_tag = card.find('h3', class_='imh-rostrum-card-title')
    
    # Extract the Price (span tag)
    price_tag = card.find('span', class_='rostrum-price')
    
    # Extract all features (li tags)
    features_elements = card.find_all('li')
    features_text = " | ".join([f.get_text(strip=True) for f in features_elements]) if features_elements else "N/A"

    # Verify we have valid data and filter out the "Included In All Plans" block
    if name_tag and price_tag:
        plan_name = name_tag.get_text(strip=True)
        if "Included" not in plan_name:
            plans_list.append({
                'Plan Name': plan_name,
                'Monthly Price': price_tag.get_text(strip=True),
                'Features': features_text
            })

# STEP 4: Display the final results
print(f"\nScraping complete! Found {len(plans_list)} plans:")
pprint.pprint(plans_list)

# driver.quit()
```

```python
import csv

# Define the file name
filename = "inmotion_hosting_plans.csv"

# Define the column headers (the keys from your dictionary)
fields = ['Plan Name', 'Monthly Price', 'Features']

# Writing to the csv file
with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
    # Create a writer object
    writer = csv.DictWriter(csvfile, fieldnames=fields)
    
    # Write the header row
    writer.writeheader()
    
    # Write all the plan data
    writer.writerows(plans_list)

print(f"\nData successfully saved to {filename}")
```

```python
# Close the current window and the browser process
driver.quit()
print("Browser closed successfully.")
```
