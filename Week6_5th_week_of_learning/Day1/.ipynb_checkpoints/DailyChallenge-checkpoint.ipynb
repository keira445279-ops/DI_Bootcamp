## 1) What is Machine Learning and why it matters for data analysts
- Machine Learning is a set of methods where a computer learns to make predictions or find patterns in data without being explicitly programmed with every rule.
- It matters for data analysts because it:
  - helps automate predictions (sales, customer churn, etc.),
  - finds hidden patterns in large datasets,
  - improves decision-making based on data â€” faster and often more accurate than manual analysis,
  - expands capabilities: clustering, segmentation, recommendations, anomaly detection.

## 2) Applications of ML across industries (examples)
- Retail / e-commerce:
  - recommendation systems (what to show a customer),
  - demand forecasting.
- Banking & finance:
  - fraud detection in transactions,
  - credit scoring.
- Healthcare:
  - aiding diagnosis from images (e.g., X-rays),
  - predicting patient complications.
- Manufacturing:
  - predictive maintenance (when to replace a part),
  - production-line optimization.
- Marketing:
  - customer segmentation and personalized campaigns,
  - predicting ad campaign response.

## 3) Differences: Supervised, Unsupervised, and Reinforcement Learning

- Supervised learning
  - Definition: the model learns from data that includes inputs and known correct answers (labels).
  - When to use: when you have historical data with a target variable.
  - Example scenario: predict whether a loan will be approved (label: yes/no), predict house price.

- Unsupervised learning
  - Definition: the model finds structure or groups in data without known answers.
  - When to use: when there are no labels but you want to understand data structure.
  - Example scenario: segment customers by behavior (clustering), detect anomalies in logs.

- Reinforcement learning
  - Definition: an agent learns to take sequential actions in an environment, receiving rewards or penalties; the goal is to maximize cumulative reward.
  - When to use: when you need to optimize a sequence of actions in a dynamic environment.
  - Example scenario: controlling a warehouse robot, real-time price optimization, learning a game strategy.
