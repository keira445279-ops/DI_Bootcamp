# Applying Inferential Statistics

### Here are the hypotheses to test:
1. Age of people who left the bank and who did not are similar. Alternative: Not similar.
2. Credit score of people who left the bank and who did not are similar. Alternative: Not similar.
3. Balance of people who left the bank and who did not are similar. Alternative: Not similar.
4. Estimated Salary of people who left the bank and who did not are similar. Alternative: Not similar.

#### The most appropriate test to analyse data here is Frequentist test.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import scipy.stats
from scipy.stats import t
from scipy.special import stdtr
from numpy.random import seed
from scipy.stats import ttest_ind
import seaborn as sns

%matplotlib inline
from matplotlib import rcParams
sns.set_style("whitegrid")
sns.set_context("poster")
```

```python
matplotlib.rcParams['figure.figsize'] = (8.0, 5.0)
```

```python
## TODO : make into a dataframe called df
df = pd.read_csv('Churn_Modelling.csv')
```

```python
## TODO : output the first 5 lines
print('The first 5 lines:')
df.head()
```

```python
## TODO : Create two separate DataFrames, `df_0` and `df_1`, to filter customers who have not exited (0) and customers who have exited (1), respectively

# customers who have not exited (0)
df_0 = df[df['Exited'] == 0] 

# customers who have exited (1)
df_1 = df[df['Exited'] == 1]
```

## Hypothesis 1: Age

```python
## TODO: Plot the age distribution for customers who stayed with the bank and those who left using seaborn, with different colors for each group and a legend.

sns.set_style("whitegrid")

plt.figure(figsize=(15,5))
sns.kdeplot(
    data=df,
    x='Age',
    hue='Exited',
    fill=True,
    common_norm=False,
    alpha=0.5,
    palette={0: 'green', 1: 'red'}
)

plt.title('Customer Age Distribution: NOT Exited vs.Exited', fontsize=20, pad = 13, fontweight='bold')
plt.xlabel('Clients Age', fontsize=15, labelpad = 10)
plt.ylabel('Density', fontsize=15, labelpad = 10)
plt.legend(title='Exited', labels=['0 (NOT Exited)', '1 (Exited)'])
plt.grid(linewidth = 0.4)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)

plt.legend(title='Exited', labels=['0 (NOT Exited)', '1 (Exited)'], fontsize = 10)
legend = plt.gca().get_legend()
legend.get_title().set_fontsize(12)

plt.show()
```

```python
## TODO: Calculate the mean and standard deviation of the age for customers who stayed with the bank.

print (f'''Customers, who STAYED with the bank
{'=' * 36}
Mean : {df_0['Age'].mean():.2f}
Standatd Deviation : {df_0['Age'].std():.2f}''')
```

```python
## TODO: Calculate the mean and standard deviation of the age for customers who left the bank.

print (f'''Customers, who LEFT the bank
{'=' * 36}
Mean : {df_1['Age'].mean():.2f}
Standatd Deviation : {df_1['Age'].std():.2f}''')
```

```python
## TODO: Perform a t-test to compare the ages of customers who stayed and left the bank.

t_statistic, p_value = ttest_ind(df_0['Age'], df_1['Age'])
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
```

---
- There are significant diferences between customers who stayed and left the bank by age feature.
- The H0 hypothesis is rejected and the alternative hypothesis (H1) is confirmed: <br>
-> Age of people who left the bank and who did not are NOT similar. 
***

### Using Bootstrapping

```python
## TODO: Write a function to perform bootstrap sampling and calculate the statistic of interest.
def bs_choice(data, func, size):
    bs_s = np.zeros(size)
    for i in range(size):
        bs_abc = np.random.choice(data, size=len(data), replace=True)
        bs_s[i] = func(bs_abc)
    return bs_s
```

```python
## TODO: Calculate the difference in means and shift the ages to the overall mean.

age_0 = df_0['Age'].values 
age_1 = df_1['Age'].values

overall_mean = np.concatenate((age_0, age_1)).mean()

observed_diff_mean = age_0.mean() - age_1.mean()

age_0_shifted = age_0 - age_0.mean() + overall_mean
age_1_shifted = age_1 - age_1.mean() + overall_mean

print (f'''Shifted ages
{'=' * 36}
LEFT : {age_0_shifted.mean():.2f}
STAYED : {age_1_shifted.mean():.2f}''')
```

```python
## TODO: Perform bootstrap sampling to calculate the standard deviation for both groups and their difference.

B = 10000

bs_std_0 = bs_choice(age_0, np.std, B)

bs_std_1 = bs_choice(age_1, np.std, B)

bs_std_diffs = bs_std_0 - bs_std_1

print(f"Mean bootstrap distribution of Std. Dev. (Group 0): {bs_std_0.mean():.4f}")
print(f"Mean bootstrap distribution of Std. Dev. (Group 1): {bs_std_1.mean():.4f}")
print(f"Mean difference of bootstrap distributions of Std. Dev.: {bs_std_diffs.mean():.4f}")
```

```python
## TODO: Calculate the p-value by comparing the difference in means to the bootstrap distribution.

bs_means_0 = bs_choice(age_0_shifted, np.mean, B)
bs_means_1 = bs_choice(age_1_shifted, np.mean, B)

bs_diffs = bs_means_0 - bs_means_1

p_value = np.sum(np.abs(bs_diffs) >= np.abs(observed_diff_mean)) / len(bs_diffs)

print(f"Observed mean difference: {observed_diff_mean: .4f}")
print(f"Bootstrap P-value: {p_value: .4f}")
```

### Conclusion
Do we reject the Null Hypothesis ? Why ?
- Yes, we reject the H0, because p-value < 0.05 and that says that there are significant diferences between customers who stayed and left the bank by age feature.
The H0 hypothesis is rejected and the alternative hypothesis (H1) is confirmed:
-> Age of people who left the bank and who did not are NOT similar.

## Hypothesis 3: Balance

```python
## TODO: Plot the distribution of Balance for both groups (Still with bank and Left the bank).

plt.figure(figsize = (15, 5))

sns.histplot(data = df, x ='Balance', hue = 'Exited', multiple='stack', palette= 'Set2', legend = True )

plt.title('Balance distribution of both groups', fontweight='bold', fontsize = 20, pad = 10)
plt.xlabel('Balance', fontsize = 15, labelpad = 10)
plt.ylabel('Quantity', fontsize = 15, labelpad = 10)
plt.grid(linewidth = 0.4)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)

plt.legend(title='Exited', labels=['0 (NOT Exited)', '1 (Exited)'], fontsize = 10)
legend = plt.gca().get_legend()
legend.get_title().set_fontsize(12)

plt.show()
```

```python
## TODO: Perform a t-test to compare the Balance between customers who stayed with the bank and those who left.

t_statistic, p_value = ttest_ind(df_0['Balance'], df_1['Balance'])
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
## TODO: Visualize the distribution of Balance for customers who stayed with the bank and those who left, excluding zero balances.

df_pos_balance = df[df['Balance'] > 0]

plt.figure(figsize = (15, 5))

sns.histplot(data = df_pos_balance, x = 'Balance' , hue = 'Exited', multiple='stack', palette= 'Set2', legend = True )

plt.title('Balance distribution of both groups', fontweight='bold', fontsize = 20, pad = 10)
plt.xlabel('Balance', fontsize = 15, labelpad = 10)
plt.ylabel('Quantity', fontsize = 15, labelpad = 10)
plt.grid(linewidth = 0.4)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)

plt.legend(title='Exited', labels=['0 (NOT Exited)', '1 (Exited)'], fontsize = 10)
legend = plt.gca().get_legend()
legend.get_title().set_fontsize(12)

plt.show()
```

```python
## TODO: Perform a t-test to compare the Balance between customers who stayed with the bank and those who left, excluding zero balances.

df_0_nonzero = df_0[df_0['Balance'] > 0]
df_1_nonzero = df_1[df_1['Balance'] > 0]

t_statistic, p_value = ttest_ind(df_0_nonzero['Balance'], df_1_nonzero['Balance'])
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
```

## Conclusion

Do we reject the Null Hypothesis ? Why ?<br>
Based on the t-test results (p = 0.17), we fail to reject the null hypothesis(Balance of people who left the bank and who did not are similar). This suggests that, after excluding zero balances, there is no statistically significant difference in the average balance between customers who stayed with the bank and those who exited.

### Using Bootstrapping

```python
## TODO: Calculate the difference in means and shift the EstimatedSalary for both groups.

estsal_0 = df_0['EstimatedSalary'].values 
estsal_1 = df_1['EstimatedSalary'].values

overall_mean = np.concatenate((estsal_0, estsal_1)).mean()

observed_diff_mean2 = estsal_0.mean() - estsal_1.mean()

estsal_0_shifted = estsal_0 - estsal_0.mean() + overall_mean
estsal_1_shifted = estsal_1 - estsal_1.mean() + overall_mean

print (f'''Shifted ages
{'=' * 36}
LEFT : {age_0_shifted.mean():.2f}
STAYED : {age_1_shifted.mean():.2f}''')
```

```python
## TODO: Calculate the bootstrap sample means for both groups and their difference.

B = 10000

bs_mean_0 = bs_choice(estsal_0, np.mean, B)

bs_mean_1 = bs_choice(estsal_1, np.mean, B)

bs_mean_diffs = bs_mean_0 - bs_mean_1

print(f"Mean bootstrap distribution (Group 0): {bs_mean_0.mean():.4f}")
print(f"Mean bootstrap distribution (Group 1): {bs_mean_1.mean():.4f}")
print(f"Mean difference of bootstrap distributions: {bs_mean_diffs.mean():.4f}")
```

```python
## TODO: Calculate the p-value based on the bootstrap distribution of the difference in means.

bs_means2_0 = bs_choice(estsal_0_shifted, np.mean, B)
bs_means2_1 = bs_choice(estsal_1_shifted, np.mean, B)

bs_diffs2 = bs_means2_0 - bs_means2_1

p_value = np.sum(np.abs(bs_diffs2) >= np.abs(observed_diff_mean2)) / len(bs_diffs2)

print(f"Observed mean difference: {observed_diff_mean: .4f}")
print(f"Bootstrap P-value: {p_value: .4f}")
```

### Conclusion
Do we reject the Null Hypothesis ? Why ?
We do not regret Null Hypothesis - Estimated Salary of people who left the bank and who did not are similar.<br>
Maybe it`s not the same, there are differences, but due to p-value (> 0.05) thay are not significant.
