### Why Nunpy and not Regular Python list?

#### Modifying the numbers in a list

```python
# Regular Python
python_list = [1, 2, 3, 4, 5]

for i in range (len(python_list)):
    python_list[i] = python_list[i] * 2

python_list
```

```python
python_list = [1, 2, 3, 4, 5]

python_list_doubled = [x * 2 for x in python_list]

python_list_doubled
```

```python
sum(python_list_doubled)
```

```python
# Numpy Array
numpy_array = np.array([1, 2, 3, 4, 5])

doubled = numpy_array * 2

doubled
```

```python
doubled.sum() # numpy has its own mathematical methods
```

```python
import time 

size = 1_000_000
python_list = list(range(size)) # creats a list in range of 1 000 000
numpy_array = np.arange(size)

start = time.time() # before running the doubling
result_list = [x * 2 for x in python_list]
python_time = time.time()- start # after running the doubling

python_time
```

```python
start = time.time() # before running the doubling
result_list = numpy_array * 2
python_time = time.time()- start # after running the doubling

python_time
```
