## Exercise 1 : Minimum and Maximum of Random Array
Create a 5x5 array with random values and find the minimum and maximum values.

```python
np.random.seed(42) # data will never change
array = np.random.uniform(1, 50, size = (5,5)).round(2) # create random numbers and round them
max_array = np.max(array) # count max
min_array = np.min(array) # count min

print(f'Min: {min_array}, Max: {max_array}') # print the result
```

## Exercise 2 : Matrix Normalization
Normalize a 3x3 random matrix (subtract the mean and divide by the standard deviation of the matrix).

```python
np.random.seed(42) # data will never change
array = np.random.uniform(25, 60, size = (3,3)).round(2) # create random numbers and round them
array
```

```python
std_array = array.std() # count std
std_array
```

```python
mean_array = array.mean() # count mean
mean_array
```

```python
norm_array = (array - mean_array) / std_array # normalize array
norm_array
```

## Exercise 4 : Matrix Multiplication
Multiply a 5x3 matrix by a 3x2 matrix (real matrix product).

```python
np.random.seed(42) # data will never change
matrix_a = np.random.uniform(1, 50, size = (5,3)).round(1)  # create random numbers and round them
matrix_a
```

```python
np.random.seed(41) # data will never change
matrix_b = np.random.uniform(1, 50, size = (3,2)).round(1)  # create random numbers and round them
matrix_b
```

```python
# Matrix multiplication
result = matrix_a @ matrix_b
result
```

```python
print(f'Matrix A: {matrix_a}')
print(f'Matrix B: {matrix_b}')
print(f'Matrix Multiplication Result: {result}')
```
