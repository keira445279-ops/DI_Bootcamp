```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import matplotlib.cm as cm
```

## 1. Data Preparation:

- Hint 1: Use np.random.uniform(low, high, size) to generate the temperature data.
- Hint 2: Create a DataFrame using pd.DataFrame(data, index, columns) with appropriate index and columns.

- Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.

- Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.- 

```python
# generate df

cities =  ['New York', 'Los Angeles', 'Washington', 'Chicago', 'Houston', 'San Diego', 'Dallas', 'Phoenix', 'San Antonio', 'Philadelphia']
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

np.random.seed (42) # tenperature will always be the same
data = np.random.uniform(low=-5, high=35, size=(len(cities), len(months))) # generate tempreture

df = pd.DataFrame(data=data, index=cities, columns=months).round(1)
df
```

## 2. Data Analysis:

- Hint 1: Calculate the annual average temperature using DataFrame.mean(axis).
- Hint 2: Find the city with the highest and lowest average temperature using idxmax() and idxmin() methods.

- Calculate the annual average temperature for each city.

- Identify the city with the highest and lowest average temperature for the year.

```python
df['aver_temp'] = df.mean(axis=1) # count average and create a new column
```

```python
df['max_temp'] = df.max(axis=1) # count max and create a new column

df['min_temp'] = df.min(axis=1) # count min and create a new column
df
```

```python
city_max = df['max_temp'].idxmax() # find a city with a maximum temperature
city_max
```

```python
city_min = df['min_temp'].idxmin() # find a city with a low temperature
city_min
```

## 3. Data Visualization

```python
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize = (20,7))

# different styles for each line
linestyles = ['-', '--', '-.', ':', '-', '--', '-.', ':', '-', '--']
markers = ['o', 's', 'D', '^', 'v', '<', '>', 'p', '*', '+']

# colors for all cities
colors = cm.tab10(np.linspace(0, 1, len(df.index)))  

for city, color, style, marker in zip(df.index, colors, linestyles, markers):
    plt.plot(months, df.loc[city, months],
             label=city, color=color, linestyle=style, marker=marker)


plt.title('Monthly Temperature Trends or Each City', fontsize = 20, pad = 15)
plt.xlabel('Month', fontsize = 13)
plt.ylabel('Temperature (C)', fontsize = 13, labelpad = 13)

plt.xticks(rotation = 90)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # legend outside the chart
plt.grid(alpha=0.3)  # light grid

plt.show()
```
