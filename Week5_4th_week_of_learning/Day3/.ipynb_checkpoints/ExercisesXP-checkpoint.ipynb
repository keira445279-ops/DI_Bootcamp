## Exercise 2: Descriptive Statistics
**Task**: Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.

```python
data = [12, 15, 13, 12, 18, 20, 22, 21]
```

```python
describe = scipy.stats.describe(data) # make variable for descriptive statistic
```

```python
# mean
mean = describe.mean

# median
median = ndimage.median(data)

# variance
variance = describe.variance

# standart deviation
std = np.sqrt(variance)

print(f'''Mean : {mean}
Median : {median}
Variance : {variance}
Standart deviation : {std}''')
```

## Exercise 3: Understanding Distributions
**Task**: Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.

```python
# create a normal distribution
group = stats.norm.rvs(loc=50, scale=10, size=50)
group
```

```python
plt.figure(figsize = (15,5))
plt.hist(group, bins = 20, color = 'mediumaquamarine', edgecolor = 'grey', linewidth=0.5)
plt.title('Normal Distribution (mean=50, std=10)', fontsize = 20, pad = 20)
plt.xlabel('Value', fontsize = 15, labelpad = 15)
plt.ylabel('Frequency', fontsize = 15, labelpad = 15)

plt.show()
```

## Exercise 4: T-Test Application
**Task**: Perform a T-test on two sets of randomly generated data.

```python
np.random.seed(36)
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_statistic, p_value = stats.ttest_ind(data1, data2) # count t-test
print(f'T-Statistic is {t_statistic.round(2)}')
print(f'P-Value is {p_value}')
```

**Conclusion**:
- T-Statistic is -7.91 
- P-Value is 1.8025562993405966e-13
- There are **significant differences**, because P-Value is lower than 0.05

## Exercise 5: Linear Regression Analysis

**Objective**: Apply linear regression to a dataset and interpret the results.

**Task**:<br>
Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), use linear regression to predict the price of a house given its size.<br>

**Dataset**:
- house_sizes: [50, 70, 80, 100, 120] (in square meters)
- house_prices: [150,000, 200,000, 210,000, 250,000, 280,000] (in currency units)<br>

**Questions**:
- What is the slope and intercept of the regression line?
- Predict the price of a house that is 90 square meters.
- Interpret the meaning of the slope in the context of housing prices.

```python
house_sizes = np.array([50, 70, 80, 100, 120]) # sqm
house_prices = np.array([150_000, 200_000, 210_000, 250_000, 280_000]) # currency units

slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices) # count line regression
print(f'''Slope is {slope},
Intercept is {intercept},
Correlation is {r_value}
P-Value is {p_value}
Standart Error is {std_err}''')
```

```python
# Predict the price of a house that is 90 square meters
price_for_90 = 1828.7671232876712 * 90 + 64383.56164383562
price_for_90
```

**Conclusions**:
- scole is 1828.7671232876712 - if house size increases by 1, than house price increases by scole. It is the angle of inclination of a line
- intercept is 64383.56164383562. It is the point where house size value and house price value are crossed <br><br>
- the cost of the house with 90 square meters will be near 229 000 (228972.60273972602)<br><br>
- so the slope is the price for 1 sqrm

## Exercise 6: Understanding ANOVA

**Objective**: Test understanding of ANOVA and its application.

**Task**:<br> 
Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.<br>

**Dataset**:
- fertilizer_1: [5, 6, 7, 6, 5]
- fertilizer_2: [7, 8, 7, 9, 8]
- fertilizer_3: [4, 5, 4, 3, 4]<br>

**Questions**:
- Perform an ANOVA test on the given data. What are the F-value and P-value?
- Based on the P-value, do the fertilizers have significantly different effects on plant growth?
- Explain what would happen if the P-value were greater than 0.05.

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f'''F-value: {f_value},
P-value: {p_value}''')
```

**Conclusions**:
- the f-value is 28.52631578947368
- the p-value is 2.754267209804844e-05<br><br>
- p-value says that the f-value is high enough to be significant differences between the groups (the avereges values of groups are significantly different and the dispersion within a group is very high). So the fertilizers have significantly different effect on plan growth<br><br>
- if the p-value are greater than 0.05 then there are no significant differences in f-value

## Exercise 7: Probability Distributions (Optional)

- Work with a binomial distribution: calculate probabilities for different numbers of successes.
- **Example**: Calculating the probability of getting exactly 5 heads in 10 coin flips.

```python
n, p = 10, 0.5  # number of trials, probability of each trial
binom_dist = stats.binom(n, p)
print("PMF at 5 successes:", binom_dist.pmf(5))
```

**Conclusion**:
if you repeat the experiment many times, in 24,6% of chances you will get exactly 5 heads in 10 coin flips

## Exercise 8: Correlation Coefficients (Optional)
- Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.
- **Example** in code: data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

```python
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
```

```python
data
```

```python
corr_p,_ = pearsonr(data['age'], data['income'])
corr_p # positive correlation, both variables increases
```

```python
corr_s,_ = spearmanr(data['age'], data['income'])
corr_s # positive correlation, both variables increases
```
