# 3. Statistical Analysis with NumPy and SciPy

---
#### I want to check two hypotheses:
- H1. The more RAM, the higher the price 
- H2. The price does not depend on the battery capacity<br><br>
That is why we will make a detailed statistical analysis for two features:
- RAM
- battery_power
---

```python
# Calculation of central tendency measures (mean, median, mode).
# Analysis of variability (range, variance, standard deviation).
# Evaluation of distribution shapes through skewness and kurtosis
```

```python
# Detailed statistical analise for RAM (MB)

# make describe variable for scipy
describe_RAM = scipy.stats.describe(train['ram'])

RAM_mean = describe_RAM.mean
RAM_median = np.median(train['ram'])
RAM_mode = stats.mode(train['ram'])[0]

RAM_range = describe_RAM.minmax
RAM_variance = describe_RAM.variance
RAM_std = np.std(train['ram'])

RAM_ske = describe_RAM.skewness
RAM_kur = describe_RAM.kurtosis

print(f'''Detailed statistical analise for RAM:
Mean : {RAM_mean}
Median : {RAM_median}
Mode : {RAM_mode}

Range : {RAM_range}
Variance : {RAM_variance:.2f}
Standard deviation : {RAM_std:.2f}

Skewness : {RAM_ske:.4f}
Kurtosis : {RAM_kur:.4f}''')
```

---
**Conclusions about RAM feature**:
- The distribution of RAM feature is around zero which is say about perfectly symmetrical, like a normal distribution, and the mean and median are almost equal.
- Krtosis is negative, which is mean that the RAM feature has thinner tails than a normal distribution, indicating a lower probability of extreme outliers.
- The RAM feature is near the normal distribution and we can use it for statistical analysis
---

```python
# Detailed statistical analise for battery_power (MA)

# make describe variable for scipy
describe_BP = scipy.stats.describe(train['battery_power'])

BP_mean = describe_BP.mean
BP_median = np.median(train['battery_power'])
BP_mode = stats.mode(train['battery_power'])[0]

BP_range = describe_BP.minmax
BP_variance = describe_BP.variance
BP_std = np.std(train['battery_power'])

BP_ske = describe_BP.skewness
BP_kur = describe_BP.kurtosis

print(f'''Detailed statistical analise for Battery Power:
Mean : {BP_mean}
Median : {BP_median}
Mode : {BP_mode}

Range : {BP_range}
Variance : {BP_variance:.2f}
Standard deviation : {BP_std:.2f}

Skewness : {BP_ske:.4f}
Kurtosis : {BP_kur:.4f}''')
```

---
**Conclusions about Battery Power feature**: 
- The situation with the Battery Power feature is the same as with RAM feature:
    - The distribution of Battery Power feature is around zero which is say about perfectly symmetrical, like a normal distribution, and the mean and median are almost equal.
    - Krtosis is negative, which is mean that the Battery Power feature has thinner tails than a normal distribution, indicating a lower probability of extreme outliers.
    - The Battery Power feature is also near the normal distribution and we can use it for statistical analysis
---

```python
# make names of the features for correlation matrix
new_names = {
    'battery_power': 'Battery Capacity(mAh)',
    'pblue': 'Bluetooth',
    'clock_speed': 'Speed of operations(GHz)',
    'dual_sim': 'Dual SIM support',
    'fc': 'Frontal camera(MP)',
    'four_g': '4G',
    'int_memory': 'Internal memory(GB)',
    'm_dep': 'Mobile Depth(cm)',
    'mobile_wt': 'Mobile Weight(grm)',
    'n_cores': 'Number of Cores(CPU)',
    'pc': 'Primary Camera(MP)', 
    'px_height': 'Vertical resolution(pxl)',
    'px_width': 'Horizontal resolution(pxl)',
    'ram': 'Random Access Memory(MB)', 
    'sc_h': 'Screen height(cm)', 
    'sc_w': 'Screen width(cm)',
    'talk_time': 'Talk time(Battery)',
    'three_g': '3G', 
    'touch_screen': 'Touch Screen',
    'wifi': 'Wi-Fi',
    'price_range': 'Price range'
}

# renaim columns
train_renamed = train.rename(columns=new_names)

# plot a correalation matrix
corr_matrix = train_renamed.corr(numeric_only=True).round(2)

# fontsize 
annot_settings = {"fontsize": 8} # fontsize inside the matrix

plt.figure(figsize=(15, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, annot_kws=annot_settings)
plt.title('Correlation Matrix')
plt.show()
```

---
**Conclusion for correlation matrix**:
- In this dataset, a mostly positive correlation is observed between the variables.
- The strongest correlation is the one between RAM (Random Access Memory) and the mobile phone's price — the higher the RAM, the higher the price.
- There are other strong correlations, but they are less interesting and do not play a significant role in the mobile phone pricing.
---

```python
# Investigate feature-target correlations using SciPy.
# As price_range feature is categorial we can`t use Pearson correlation
# But we can use Spearman correlation

#Spearman correlation for RAM
spearman_RAM = spearmanr(train['ram'], train['price_range'])

#Spearman correlation for Battery_power
spearman_BP = spearmanr(train['battery_power'], train ['price_range'])

print (f'''Spearman Correlation:
For RAM : {spearman_RAM}
For Battery Power : {spearman_BP}''')
```

------------------------------------------------------------------------------------------------------------------
**Correlation conclusions**:
- Both correlations are positive and statistically significant.
- However, RAM has a much stronger and likely more significant relationship with Price Range than Battery Power.
---

```python
# Perform hypothesis testing for statistical significance between groups (e.g., different price ranges).
# We will perform it with Anova

# Anova for RAM
groups = [train[train['price_range']==i]['ram'] for i in range(4)]

f_value, p_value = stats.f_oneway(*groups)

print (f'''Anova Results for RAM:
F-value : {f_value:.2f}
P-value : {p_value}''')
```

---
**! The 1st hypothesis  - The more RAM, the higher the price - is confirmed**
***

```python
# Calculate the distribution between two price categories for RAM 
from scipy.stats import mannwhitneyu

# criteria of Mann-Whitneyu for RAM
mann_ram_01 = mannwhitneyu(
    train[train['price_range']==0]['ram'],
    train[train['price_range']==1]['ram']
)

mann_ram_12 = mannwhitneyu(
    train[train['price_range']==1]['ram'],
    train[train['price_range']==2]['ram']
)

mann_ram_23 = mannwhitneyu(
    train[train['price_range']==2]['ram'],
    train[train['price_range']==3]['ram']
)

print (f'''Mann-Whitneyu for RAM:
For 0-1 Price Range : {mann_ram_01}
For 1-2 Price Range : {mann_ram_12}
For 2-3 Price Range : {mann_ram_23}''')
```

---
**Conclusion about results of Mann-Whitneyu for RAM**:
- The distribution (or median) of RAM is statistically significantly different across the compared groups.
- The amount of RAM is highly dependent on the price range of the device.
---

```python
# Plot Boxplot for RAM

# Set the figure size for better readability
plt.figure(figsize=(15, 6))

# Generate the Box Plot
sns.boxplot(
    x='price_range', # Categorical variable (X-axis)
    y='ram',         # Quantitative variable (Y-axis)
    data=train,
    palette='viridis', # A nice color scheme
    hue = 'price_range'
)

ax = plt.gca()
ax.set_xticklabels(['Low Cost', 'Medium Cost', 'High Cost', 'Very High Cost'])

# remove the legend
if ax.legend_ is not None:
    ax.legend_.remove()
    
# Set the labels and title for the plot
plt.xlabel('Price Range', fontsize=12, labelpad = 8)
plt.ylabel('RAM (Random Access Memory) in MB', fontsize=12)
plt.title('RAM Distribution Across Phone Price Categories', fontsize=14, pad = 10)

# Add a grid for easier reading of the Y-axis values
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
# Anova for Battery Power
groups = [train[train['price_range']==i]['battery_power'] for i in range(4)]

f_value, p_value = stats.f_oneway(*groups)

print (f'''Anova Results for Battery Power:
F-value : {f_value:.2f}
P-value : {p_value:.4f}''')
```

---
**! The 2nd hypothesis - The price does not depend on the battery capacity - is NOT confirmed**

Price is DEPEND on the batery power
***

```python
# Calculate the distribution between two price categories for Battery Power
# criteria of Mann-Whitneyu for Battery Power
mann_bp_01 = mannwhitneyu(
    train[train['price_range']==0]['battery_power'],
    train[train['price_range']==1]['battery_power']
)

mann_bp_12 = mannwhitneyu(
    train[train['price_range']==1]['battery_power'],
    train[train['price_range']==2]['battery_power']
)

mann_bp_23 = mannwhitneyu(
    train[train['price_range']==2]['battery_power'],
    train[train['price_range']==3]['battery_power']
)

print (f'''Mann-Whitneyu for Battery Power:
For 0-1 Price Range : {mann_bp_01}
For 1-2 Price Range : {mann_bp_12}
For 2-3 Price Range : {mann_bp_23}''')
```

---
**Conclusion about results of Mann-Whitneyu for Battery Power**:
- Battery power is a significant factor when transitioning from the lowest price range (0-1) to the highest price range (2-3).
- However, differences in battery power are not definitive or significant for separating the 1-2 price range and the middle price range (1-2). It is possible that in this price segment, other characteristics (camera, design, brand) play a more important role than battery power.
---

```python
# Plot Boxplot for Battery Power

# Set the figure size for better readability
plt.figure(figsize=(15, 6))

# Generate the Box Plot
sns.boxplot(
    x='price_range', # Categorical variable (X-axis)
    y='battery_power',         # Quantitative variable (Y-axis)
    data=train,
    palette='viridis', # A nice color scheme
    hue = 'price_range'
)

ax = plt.gca()
ax.set_xticklabels(['Low Cost', 'Medium Cost', 'High Cost', 'Very High Cost'])

# remove the legend
if ax.legend_ is not None:
    ax.legend_.remove()
    
# Set the labels and title for the plot
plt.xlabel('Price Range', fontsize=12, labelpad = 8)
plt.ylabel('Battery Power', fontsize=12)
plt.title('Battery Power Distribution Across Phone Price Categories', fontsize=14, pad = 10)

# Add a grid for easier reading of the Y-axis values
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

---
## Overall conclusions:
- The analysis provides a comprehensive understanding of how key technical characteristics influence the price range of mobile phones in the dataset.
- Both RAM and Battery Power exhibit distributions close to normal, with symmetric shapes, nearly identical mean and median values, and negative kurtosis. This indicates thinner-than-normal tails and a low likelihood of extreme outliers, suggesting that both features are suitable for statistical testing and modeling.
- Correlation analysis reveals that most features have positive relationships with the target variable, but RAM stands out as the strongest predictor of price. Higher RAM values are consistently associated with higher price ranges, and this relationship is statistically significant across all tests performed. This fully confirms the first hypothesis: devices with more RAM tend to belong to higher price categories.
- Battery Power shows a weaker overall correlation with price, but the statistical tests reveal meaningful differences between the lowest and highest price groups. While battery capacity is not a decisive factor in the mid-range price segment, it does play a role when distinguishing budget devices from premium ones. Therefore, the second hypothesis — that battery capacity does not affect price — is rejected. Battery Power does influence price, but its impact is more noticeable at the extremes of the pricing spectrum and less relevant for intermediate categories, where other features likely dominate.
- **In summary**, RAM is a strong and consistent predictor of mobile phone price, while Battery Power has a moderate, context-dependent influence, contributing mainly to the differentiation between low-end and high-end devices.
