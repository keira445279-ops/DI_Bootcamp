## Exercise 1 : Matrix Operations

**Instructions**<br>
In this exercise, you’ll work with a 3x3 matrix. Here’s a brief explanation of the concepts:

**Determinant**: The determinant is a value that can be computed from the elements of a square matrix. It provides important information about the matrix, such as whether it has an inverse, and is used in various areas like linear algebra and calculus. To understand more about it you can watch this video.<br>
**Inverse of a Matrix**: The inverse of a matrix is a matrix that, when multiplied with the original matrix, results in an identity matrix. Not all matrices have inverses. The inverse is crucial in solving systems of linear equations.

- Create a 3x3 matrix and perform the following operations:
- Calculate the determinant.
- Find the inverse of the matrix.

```python
np.random.seed(30) # numbers will never change
matrix = np.random.randint(10, 20, size = (3,3)) # create matrix
matrix
```

```python
# finding determinant

deter = np.linalg.det(matrix)
deter
```

```python
# inverse of a matrix

inv = np.linalg.inv(matrix) 
inv
```

## Exercise 2 : Statistical Analysis

**Instructions**<br>
In this exercise, you’ll calculate statistical measures for a dataset:

- **Mean**: The average value of a dataset.
- **Median**: The middle value in a dataset when it is arranged in ascending or descending order.
- **Standard Deviation**: A measure of the amount of variation or dispersion in a set of values.

Using NumPy, generate an array of 50 random numbers and compute:
- The mean and median.
- The standard deviation.

```python
np.random.seed(30) # numbers will never change
# create array and round it
array = np.random.uniform(10, 70, size = 50).round(1)
```

```python
# count mean
array_mean = array.mean().round(1)

# count median
array_median = np.median(array).round(1)

# count standart deviation
array_std = array.std().round(1)

# print the result
print(f'''Mean : {array_mean} 
Median : {array_median} 
Standart Deviation : {array_std}''')
```

## Exercise 4 : Data Manipulation with NumPy and Pandas

**Instructions**<br>
Create a DataFrame with random numbers and perform:
- Conditional selection of data.
- Aggregation functions like sum and average.

```python
# create df

cities = ['New York', 'Milan', 'Oslo', 'Magadan', 'Prague', 'Riga', 'Krakow']
sum_months = ['Jun', 'Jul', 'Aug']

np.random.seed(42)
data = np.random.uniform(22, 33, size = (len(cities), len(sum_months))).round(1)

temp = pd.DataFrame(data = data , index = cities, columns = sum_months)
temp
```

```python
# conditional selection of data

high_temp = temp['Jul'] > 30
low_temp = (temp['Jun'] < 30) & (temp['Aug'] < 29)

temp[high_temp], temp[low_temp]
```

```python
# agregate sum for needed columns
temp['temp_sum'] = temp.loc[:, sum_months].sum(axis = 1)

# agregate average for needed columns
temp['temp_avr'] = temp.loc[:, sum_months].mean(axis = 1).round(1)
temp
```

## Exercise 6 : Basic Hypothesis Testing

**Instructions**
- Create a sample dataset to test the effectiveness of a new training program on employee productivity:
- Given a dataset, formulate a simple hypothesis and test it using basic statistical functions in NumPy.

```python
# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)


# H1 - new training program is effective to increase employee productivity
```

```python
# find meadian and mean BEFORE
pb_median = np.median(productivity_before).round(1)
pb_mean = productivity_before.mean().round(1)

# find std BEFORE
pb_std = productivity_before.std().round(1)

# find min and max BEFORE
pb_min = productivity_before.min().round(1)
pb_max = productivity_before.max().round(1)

# find percentiles BEFORE
pb_perc = np.percentile(productivity_before, [25, 50, 75]).round(1)
```

```python
# find meadian and mean AFTER
pa_median = np.median(productivity_after).round(1)
pa_mean = productivity_after.mean().round(1)

# find std AFTER
pa_std = productivity_after.std().round(1)

# find min and max AFTER
pa_min = productivity_after.min().round(1)
pa_max = productivity_after.max().round(1)

# find percentiles AFTER
pa_perc = np.percentile(productivity_after, [25, 50, 75]).round(1)
```

```python
print (f'Median before : {pb_median} | Mediane after : {pa_median}')
print (f'Mean before : {pb_mean} | Mean after : {pa_mean}')
print (f'Std before : {pb_std} | Std after : {pa_std}')
print (f'Min before : {pb_min} | Min after : {pa_min}')
print (f'Max before : {pb_max} | Max after : {pa_max}')
print (f'Percentiles before : {pb_perc} | Percentiles after : {pa_perc}')
```

```python
improvement = productivity_after - productivity_before
mean_improvement = np.mean(improvement).round(1)
median_improvement = np.median(improvement).round(1)

print (f'For each enployee average increase of productivity after training is {mean_improvement} and the median increase is {median_improvement}')
```

```python
# After training the productivity of employees increased (mean, median and standart deviation became higher). 
# Also we this in average productivity level for every employee, it was increase in averege by 4.7 
```
