```python
# drop columns with > 90% of missing data

cols_to_drop = [
    'other_fuel1', 'other_fuel2', 'other_fuel3',
    'generation_gwh_2013','generation_gwh_2014','generation_gwh_2015',
    'generation_gwh_2016','generation_gwh_2017','generation_gwh_2018','generation_gwh_2019',
    'generation_data_source',
    'estimated_generation_gwh_2013','estimated_generation_gwh_2014',
    'estimated_generation_gwh_2015','estimated_generation_gwh_2016', 'wepp_id'
]
plants = plants.drop(columns=cols_to_drop)
```

```python
# fillna columns with 10% - 20% of missing data

plants['source'] = plants['source'].fillna('Unknown')
plants['url'] = plants['url'].fillna('Unknown')
plants['geolocation_source'] = plants['geolocation_source'].fillna(plants['geolocation_source'].mode()[0])
plants['estimated_generation_gwh_2017'] = plants['estimated_generation_gwh_2017'].fillna(plants['estimated_generation_gwh_2017'].median())
```

# 3. Statistical Analysis:

- Perform a statistical analysis of power output by fuel type using NumPy’s statistical functions.
- Use hypothesis testing to determine if the mean power output differs significantly between different fuel types.

```python
plants.info()
```

```python
# statistical numpy functions 
numpy_funcs = {
    'Count': len,             
    'Mean': np.mean,
    'Std': np.std,
    'Min': np.min,
    'Max': np.max
}
```

```python
def get_quantile(data, q):
    return np.quantile(data.dropna(), q)

#aggregate numpy functions
results = plants.groupby('primary_fuel')['estimated_generation_gwh_2017'].agg(
    Count='size',               
    Mean=np.mean,
    Std=np.std,
    Min=np.min,
    Max=np.max,
    Q25=lambda x: get_quantile(x, 0.25),
    Median=lambda x: get_quantile(x, 0.50),
    Q75=lambda x: get_quantile(x, 0.75)
).round(1)


results = results.rename(columns={'Q25': '25%', 'Median': '50%', 'Q75': '75%'})

print(results)
```

---
We will use the hypothethis that:<br>
**At least one mean value of power output differs significantly between different fuel types.**<br><br>
And we will check it with ANOVA test
***

```python
# As we have 15 fuel types, we will use ANOVA 

fuel_types = plants['primary_fuel'].unique()

groups = [plants[plants['primary_fuel'] == fuel]['estimated_generation_gwh_2017'].dropna()
          for fuel in fuel_types]

f_value, p_value = stats.f_oneway(*groups)

print (f'''Anova Results for RAM:
F-value : {f_value:.2f}
P-value : {p_value}''')
```

```python
# Plot Boxplot for fuel types

# Set the figure size for better readability
plt.figure(figsize=(20, 8))

# Generate the Box Plot
sns.boxplot(
    x='primary_fuel', # Categorical variable (X-axis)
    y='estimated_generation_gwh_2017', # Quantitative variable (Y-axis)
    data=plants,
    palette='viridis', # A nice color scheme
    hue = 'primary_fuel'
)

# ax = plt.gca()
# ax.set_xticklabels(['Low Cost', 'Medium Cost', 'High Cost', 'Very High Cost'])

# remove the legend
# if ax.legend_ is not None:
    # ax.legend_.remove()
    
# Set the labels and title for the plot
plt.xlabel('Fuel Type', fontsize=12, labelpad = 8)
plt.ylabel('Power Output(gwh)', fontsize=12)
plt.title('Power Output Distribution Across Fuel Types', fontsize=14, pad = 10)

# Add a grid for easier reading of the Y-axis values
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

# 4. Time Series Analysis

```python
# Drop NaN and make dateset copy
plants_clean = plants.dropna(subset=['commissioning_year']).copy()

# Drop NaN and make dateset copy
plants_clean['commissioning_year'] = plants_clean['commissioning_year'].astype(int)

# Filter incorrect years (e.g., 0 or future years, if there are errors in the data)
plants_clean = plants_clean[(plants_clean['commissioning_year'] > 1900) & (plants_clean['commissioning_year'] <= 2024)]

# Calculate the total capacity for each fuel type
top_fuels = plants_clean.groupby('primary_fuel')['capacity_mw'].sum().sort_values(ascending=False)

# Select the TOP-6 names (the rest will go to "Other")
# You can change the number 6 to the desired amount
main_fuels = top_fuels.head(6).index.tolist()

# Create a copy of the column so as not to corrupt the original data
plants_clean['fuel_grouped'] = plants_clean['primary_fuel'].apply(lambda x: x if x in main_fuels else 'Other')

# Build the pivot table using the new 'fuel_grouped' column
evolution_matrix = plants_clean.pivot_table(
    index='commissioning_year', 
    columns='fuel_grouped',  # Use the grouped column
    values='capacity_mw', 
    aggfunc='sum',
    fill_value=0
)

# Plot the chart (there will be only 7 colors, easy to read)
evolution_matrix.plot(kind='area', stacked=True, figsize=(14, 7), alpha=0.8)
plt.title('Generation Dynamics (TOP-6 Sources + Other)', pad = 15, fontsize = 20)
plt.ylabel('Capacity (MW)', labelpad = 7, fontsize = 15)
plt.xlabel('Commissioning Year', labelpad = 7, fontsize = 15) # Added x-axis label for clarity
plt.show()
```

---
The peak of development of electric power occurred in 2007 - 2008 years, the highest development occurred for wind and oil
***

# 6. Matrix Operations in Real-World Context

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# --- 1. PREPARATION AND STANDARDIZATION ---
features = ['capacity_mw', 'latitude', 'longitude']
X = plants[features].copy()

# standartized matrix X (X_scaled)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 2. MATRIX OPERATION: PCA EXECUTION ---
# PCA initialization: looking for 3 Principal Components (by number of features)
pca = PCA(n_components=3)

# Application of PCA: Here the covariance matrix is decomposed
pca.fit(X_scaled)

# --- 3. OUTPUT AND ANALYSIS ---
# Eigenvalues: the proportion of variance explained
print("Explained Variance Ratio:\n", pca.explained_variance_ratio_)

# Eigenvectors: Component loadings
eigenvectors = pd.DataFrame(
    pca.components_, 
    index=['PC1', 'PC2', 'PC3'], 
    columns=features
)
print("\nEigenvectors (Component Loadings):\n", eigenvectors)
```

To identify the primary factors contributing to the variance among power plants, Principal Component Analysis (PCA) was applied to the standardized attributes (capacity_mw, latitude, longitude).
- **Dimensionality Reduction**: The first two Principal Components (PC1 and PC2) account for 72.05% of the total data variance. This allows for an effective reduction in the dimensionality of the analysis from three original attributes to two synthetic factors.
- **Factor Interpretation (Eigenvectors)**:
    - PC1 (39.05% of variance): Represents a composite factor determined by capacity and longitude. This factor describes variations related to the scale of the plant and its East-West location.
    - PC2 (33.00% of variance): Is almost entirely determined by latitude.
    
This factor describes variations related exclusively to North-South location.Thus, the main factors explaining the differences in the data are combined scale/eastern location and North-South location.

# 7. Integrating NumPy with Pandas and Matplotlib

```python
# Condition 1: High power and traditional fuel
conditions = [
    (plants_clean['capacity_mw'] > 2000) & (plants_clean['primary_fuel'].isin(['Coal', 'Nuclear', 'Hydro'])),
    (plants_clean['capacity_mw'] < 100)
]

# Corresponding values
choices = [
    'Strategic Base Load',
    'Micro/Local Scale'
]

# Using np.select to create a new column
plants_clean['strategic_scale'] = np.select(
    conditions,
    choices,
    default='Standard Scale'
)

# Testing the distribution of a new factor
print("Distribution of the new 'strategic_scale' factor:")
print(plants_clean['strategic_scale'].value_counts())
```

```python

year_data = plants_clean['commissioning_year'].values

# Using np.histogram to calculate frequencies and interval boundaries
# density=True normalizes frequencies so that the total area is 1
counts, bin_edges = np.histogram(year_data, bins=30, density=True) 

plt.figure(figsize=(12, 6))

plt.hist(
    year_data, 
    bins=30, 
    density=True,
    alpha=0.6, 
    color='teal'
)

from scipy.stats import gaussian_kde 
x_range = np.linspace(year_data.min(), year_data.max(), 100)
kde = gaussian_kde(year_data)
plt.plot(x_range, kde.evaluate(x_range), color='red', linewidth=2, label='KDE Density')

plt.title('Normalized Distribution of Commissioning Year')
plt.xlabel('Commissioning Year')
plt.ylabel('Density (Proportion per Year)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

These examples show how NumPy acts as a high-performance mathematical kernel, making operations in Pandas more efficient and visualizations in Matplotlib more accurate and informative.
