### Percentiles

The value below which a given percentage of data falls.

- 25th percentile (Q1) : 25% of data is below this
- 50th percintile (Q2) : Same as median
- 75th percintile (Q3) : 75% of data is below this

Why it matters:
- Understand the data distribution: see where most data is
- Detect outliers: Values beyond 95th percentile
- Performence metrics: Top 10% of students

np.percentile(array, [25, 50, 75])

### Variance and Standart Deviation

Problem example:
- Group A: 48, 49, 50, 51, 52 -> scores are close together
- Group B: 20, 40, 50, 60, 80 -> scores are apart



```python
group_a = np.array([48, 49, 50, 51, 52])
group_b = np.array([20, 40, 50, 60, 80])

group_a.mean(), group_b.mean()
```

```python
# Variance
## calculating variance:
## 1. Find how far each values if from the mean = calculate the distance
[2, 1, 0, 1, 2]
## 2. Squares the distances (make them positive)
[4, 1, 0, 1, 4]
## 3. Average the squared distance
mean([4, 1, 0, 1, 4]) -> 2

small varaince (2) - data points are close together
```

```python
# Standart Deviation

Square root of the variance

Variance: 2_500_000_000 dollars^2
Standart deviation: 50_000 <- much clearer!

house prices typically vary by +- 50_000 from the mean
```

```python
np.array([70, 75, 80, 85, 90])
mean: 80

[10, 5, 0, 5, 10]
>
[100, 25, 0, 25, 100]
>
50 (variance, not readable)
>
7 (standart deviation, scores vary by +- 7 points) - корень из 50
```

### Correlation

Measures how two variables move together.

Correlation Coefficient (r):
- +1.0: Perfect positive correlation (both increase together)
- 0.0: No correlation (independent variables)
- -1.0: Perfect nrgative correlation (one increases, other decreases)

real world examples:
- bill size vs tip amount: positive (bigger bill -> bigger tip)
- shoe size vs IQ: Zero (unrelated)
- tenperature vs heating cost: negative (warmer -> less heating)

## Linear Algebra Operations Introduction

```python
data = np.array([
    [10, 1000],
    [20, 2000]
]) # Problem: columns 0 and 1 are completely dependant

10 * 100 = 1000
1000 // 100 = 10

# linalg = linear akgebra module in numpy
# linalog.det = determinant method (useful for detemining broken data / when we have duplicate columns)

np.linalg.det(data)
# we get 0 which means that we have a problem i.e. the columns are redundant
```

```python
data_fixed = np.array([
    [10, 50],
    [20, 30]
])

np.linalg.det(data_fixed)
# we get a number which is not 0 which means we have good data i.e. the columns are independent
```

```python
# House data
# Column 1: Square feet
# Column 2: Number of rooms (NOT derived from the square feet!)

house_data = np.array([
    [1000, 3],
    [1500, 5]
])

# Check if the data is healthy
np.linalg.det(house_data)
```
