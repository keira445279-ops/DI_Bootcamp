```python
!pip install gpt4all
from gpt4all import GPT4All

# Updated model name to a commonly available GPT4All model
# Trying another widely available model after previous 404 errors.
model = GPT4All("mistral-7b-openorca.gguf2.Q4_0.gguf")
```

```python
input_text = """
Employees must ensure that all remote access to internal systems is established via the approved secure VPN.
Under no circumstances should unsecured connections or personal devices lacking endpoint protection be used to access proprietary data or sensitive communications.
"""

# Prompt для Step 1
prompt_friendly = f"""
Rewrite the following policy text in a friendly, clear tone for employees:
- Use bullet points
- Paraphrase only (no direct quotes)
- Stay under 75 words
- Only include information from the text below

Input text:
{input_text}
"""

# The answer
response = model.generate(prompt_friendly)
print("=== Friendly Version ===")
print(response)
```

```python
prompt_no_halluc = f"""
Rewrite the text in bullet points, friendly tone.
- Only use information from the input text.
- Do not add new technologies, rules, or recommendations.
- Paraphrase only
- Keep under 75 words

Input text:
{input_text}
"""

response_clean = model.generate(prompt_no_halluc)
print(response_clean)
```

```python
prompt_intern = f"""
Paraphrase the policy for a junior intern:
- Use simple words
- Maximum 4 bullet points
- Friendly, supportive tone
- Only use information from the input text

Input text:
{input_text}
"""

response_intern = model.generate(prompt_intern)
print("=== Intern Version ===")
print(response_intern)
```

```python
prompt_quote = f"""
Extract one direct quote from the following text that captures the main security rule. Do not paraphrase.

Input text:
{input_text}
"""

response_quote = model.generate(prompt_quote)
print("=== Quote ===")
print(response_quote)
```

- In what kind of internal communication would quoting be more appropriate than paraphrasing?<br>
Quotations are useful for official communications, reminders, and legal notices.
- When might quoting pose a risk?<br>
If the text is too complex, employees may not understand or misinterpret the rules.
